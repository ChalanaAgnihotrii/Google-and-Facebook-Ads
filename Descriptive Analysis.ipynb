```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_excel('SampleData.xlsx', sheet_name='Google Raw')
df2 = pd.read_excel('SampleData.xlsx', sheet_name='Facebook Raw')
```

```python
def descriptive_analysis(df, platform):
    # Summary statistics
    print(f"\nSummary Statistics - {platform} Ads Data:")
    print(df.describe())

    # Total impressions, clicks, conversions, cost
    total_impressions = df['Impressions'].sum()
    total_clicks = df['Clicks (Traffic)'].sum()
    total_conversions = df['Traffic to Lead'].sum()
    total_cost = df['Cost (INR)'].sum()
    print(f"\nTotal Impressions: {total_impressions}")
    print(f"Total Clicks: {total_clicks}")
    print(f"Total Conversions: {total_conversions}")
    print(f"Total Cost: {total_cost}")

    # CTR (Click-through Rate)
    df['CTR'] = (df['Clicks (Traffic)'] / df['Impressions']) * 100
    average_CTR = df['CTR'].mean()
    print(f"\nAverage CTR: {average_CTR:.2f}%")

    # Cost per Click (CPC)
    df['CPC'] = df['Cost (INR)'] / df['Clicks (Traffic)']
    average_CPC = df['CPC'].mean()
    print(f"Average CPC: {average_CPC:.2f}")

    # Conversion Rate
    df['Traffic to Lead'] = (df['Traffic to Lead'] / df['Clicks (Traffic)']) * 100
    average_conversion_rate = df['Traffic to Lead'].mean()
    print(f"Average Conversion Rate: {average_conversion_rate:.2f}%")

    # Visualize CTR and Conversion Rate
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    plt.hist(df['CTR'], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'{platform} Ads CTR Distribution')
    plt.xlabel('CTR')
    plt.ylabel('Frequency')
    plt.subplot(1, 2, 2)
    plt.hist(df['Traffic to Lead'], bins=20, color='lightgreen', edgecolor='black')
    plt.title(f'{platform} Ads Conversion Rate Distribution')
    plt.xlabel('Traffic to Lead')
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

```python
descriptive_analysis(df1, 'Google')
```

```python
descriptive_analysis(df2, 'Facebook')
```
