# Data Cleaning & Data Preparation

Identifying any missing values in the datasets.

```python
def clean_and_prepare_data(df):
    # Fill missing values with mean
    df.fillna(df.mean(), inplace=True)
    # Convert date columns to datetime format
    df['Date'] = pd.to_datetime(df['Date'])
    # Remove duplicate rows
    df.drop_duplicates(inplace=True)
    # Drop unnecessary columns
    df.drop(['unnecessary_column1', 'unnecessary_column2'], axis=1, inplace=True)
    # Rename columns for consistency
    df.rename(columns={'old_column_name': 'new_column_name'}, inplace=True)
    # Handle outliers
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
    # Convert non-numeric values to numeric
    df = df.apply(pd.to_numeric, errors='coerce')
    # Drop rows with NaN values
    df.dropna(inplace=True)
```

```python
# Load Google Raw data

# Display the first few rows of Google Raw data
print(df1.head())

# Perform data cleaning and preparation for Google Raw data
clean_and_prepare_data(df1)

# Display summary statistics for Google Raw data
print(df1.describe())

# Display information about Google Raw data
print(df1.info())
```

```python
# Display the first few rows of Facebook Raw data
print("\nFacebook Raw Data:")
print(df2.head())

# Perform data cleaning and preparation for Facebook Raw data
clean_and_prepare_data(df2)

# Display summary statistics for Facebook Raw data
print(df2.describe())

# Display information about Facebook Raw data
print(df2.info())
```
